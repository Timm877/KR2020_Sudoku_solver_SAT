```python
import numpy as np 
```

```python
for hr in df['heuristic'].unique():
#     print(hr)
    _df = df[df['heuristic'] == hr].sort_values(by='avg_human_solve_time').reset_index()
#     _df = _df[0:17]
    _df['n_iterations'] = _df['n_iterations']/_df['n_iterations'].max()
    hr_values = _df['n_iterations'].values
    h_values = _df['avg_human_solve_time'].values
    _sum = np.cov([hr_values, h_values])
    _df.plot(y=['n_iterations','avg_human_solve_time'], style='o', figsize=(12,10), title=f'{hr}_{_sum}')

#     print(sum(_df['n_iteration']-_df['avg_human_solve_time']))
```
